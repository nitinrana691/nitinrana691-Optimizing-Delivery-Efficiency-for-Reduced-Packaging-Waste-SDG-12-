```python
plt.figure(figsize=(10, 6))
plt.plot(df['order_purchase_timestamp'], label='Purchase Timestamp')
plt.plot(df['order_delivered_customer_date'], label='Delivered to Customer Date')
plt.plot(df['order_delivered_carrier_date'], label='Delivered to Carrier Date')
plt.xlabel('Order Index')
plt.ylabel('Date')
plt.title('Order Purchase, Delivery to Customer, and Delivery to Carrier Dates')
plt.legend()
plt.show()
```

```python
orders_over_time = df.groupby([pd.Grouper(key='order_purchase_timestamp', freq='M'), 'order_status'])['order_id'].count().reset_index()
```

```python
plt.figure(figsize=(12, 6))
for status in orders_over_time['order_status'].unique():
  df_subset = orders_over_time[orders_over_time['order_status'] == status]
  plt.plot(df_subset['order_purchase_timestamp'], df_subset['order_id'], label=status)
plt.xlabel('Order Purchase Timestamp')
plt.ylabel('Number of Orders')
plt.title('Number of Orders Over Time by Order Status')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.fill_between(range(len(df)), df['order_purchase_timestamp'], label='Purchase Timestamp')
plt.fill_between(range(len(df)), df['order_delivered_customer_date'], label='Delivered to Customer Date')
plt.fill_between(range(len(df)), df['order_delivered_carrier_date'], label='Delivered to Carrier Date')
plt.xlabel('Order Index')
plt.ylabel('Date')
plt.title('Order Purchase, Delivery to Customer, and Delivery to Carrier Dates')
plt.legend()
plt.show()
```

```python
df_new = pd.get_dummies(df, columns=['order_status'])

df_new.head()
```

```python
df_new.info()
```

**convert the order stats into interger insted of string**

```python
df_new['order_status_approved'] = df_new['order_status_approved'].astype(int)
df_new['order_status_canceled'] = df_new['order_status_canceled'].astype(int)
df_new['order_status_created'] = df_new['order_status_created'].astype(int)
df_new['order_status_delivered'] = df_new['order_status_delivered'].astype(int)
df_new['order_status_invoiced'] = df_new['order_status_invoiced'].astype(int)
df_new['order_status_processing'] = df_new['order_status_processing'].astype(int)
df_new['order_status_shipped'] = df_new['order_status_shipped'].astype(int)
df_new['order_status_unavailable'] = df_new['order_status_unavailable'].astype(int)
```
