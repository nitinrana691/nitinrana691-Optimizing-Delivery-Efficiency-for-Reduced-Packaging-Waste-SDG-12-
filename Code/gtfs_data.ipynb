```python
# Ensure data quality by handling missing or anomalous values
dataset['SRI'] = pd.to_numeric(dataset['SRI'], errors='coerce')
dataset['SRI'].fillna(dataset['SRI'].median(), inplace=True)
dataset['speed'].replace([np.inf, -np.inf], np.nan, inplace=True)
dataset['speed'].fillna(dataset['speed'].median(), inplace=True)

# Remove outliers where speed is greater than 100
dataset = dataset[dataset['speed'] <= 100]

# Check the actual column names in your dataset
print(dataset.columns)

# Include 'Number_of_trips' in the aggregation, adjusting column names if necessary
optimal_routes = dataset.groupby('trip_id').agg({
    'time': 'sum',
    'speed': 'mean',
    'Number_of_trips': 'sum'
}).reset_index()

# Sort by travel time
optimal_routes = optimal_routes.sort_values(by='time')

print("Recommendations for optimizing routes based on travel time:")
print(optimal_routes.head())

# Suggest reducing packaging size for trips with high travel time (above 75th percentile)
high_travel_time_threshold = np.percentile(optimal_routes['time'], 75)
high_travel_time_trips = optimal_routes[optimal_routes['time'] > high_travel_time_threshold]

print("Consider reducing packaging size for trips with high travel time:")
print(high_travel_time_trips)

# Generate actionable insights and recommendations
optimal_routes_summary = optimal_routes.head(10)
high_travel_time_summary = high_travel_time_trips.head(10)

print("Top 10 Optimal Routes:")
print(optimal_routes_summary)

print("Top 10 High Travel Time Trips for Packaging Optimization:")
print(high_travel_time_summary)
# Prepare a summary DataFrame for the optimal routes
optimal_routes_summary.to_csv('optimal_routes_summary.csv', index=False)
# Save the high travel time trips to a CSV file
high_travel_time_trips.to_csv('high_travel_time_trips.csv', index=False)
```

```python
#6. Assess Environmental Impact
#Evaluate how your recommendations might impact the environment:
# Example: Estimate reduction in packaging waste
average_package_waste_per_trip = 1 # Assuming an average of 1 unit of waste per trip. Replace with actual data if available.
current_packaging_waste = dataset['Number_of_trips'].sum() * average_package_waste_per_trip

# Include 'number_of_trips' in the aggregation
optimal_routes = dataset.groupby('trip_id').agg({'time': 'sum', 'speed': 'mean', 'Number_of_trips': 'sum'}).reset_index()
optimal_routes = optimal_routes.sort_values(by='time')

high_travel_time_trips = optimal_routes[optimal_routes['time'] > optimal_routes['time'].median()]
optimized_packaging_waste = high_travel_time_trips['Number_of_trips'].sum() * average_package_waste_per_trip

waste_reduction = current_packaging_waste - optimized_packaging_waste
print(f"Estimated reduction in packaging waste: {waste_reduction}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/cleaned_dataset.csv'
gtfs_data = pd.read_csv(file_path)

# Check the actual column names
print(gtfs_data.columns)

# Replace infinite values with NaN and then drop rows with NaN values
gtfs_data.replace([np.inf, -np.inf], np.nan, inplace=True)

# Ensure 'speed' and 'sri' columns are numeric
gtfs_data['speed'] = pd.to_numeric(gtfs_data['speed'], errors='coerce')
gtfs_data['SRI'] = pd.to_numeric(gtfs_data['SRI'], errors='coerce')

# Drop rows with NaN values in the specified columns
cleaned_data = gtfs_data.dropna(subset=['speed', 'SRI'])

# Compute the correlation between speed and sri
correlation = cleaned_data[['speed', 'SRI']].corr()

# Plot the relationship between speed and sri
plt.figure(figsize=(10, 6))
plt.scatter(cleaned_data['SRI'], cleaned_data['speed'], alpha=0.5, color='purple')
plt.title('Relationship between Speed and SRI')
plt.xlabel('SRI')
plt.ylabel('Speed')
plt.grid(True)
plt.show()

# Print the correlation
print("Correlation between Speed and SRI:")
print(correlation)
```
