```python
df= pd.read_csv("/content/clean_data.csv")
# Calculate item and total volumes
df['item_volume_cm3'] = df['height__cm'] * df['width__cm'] * df['depth__cm']
df['total_item_volume_cm3'] = df['item_volume_cm3'] * df['amount']

# Assuming packaging volume is standardized and box dimensions are needed
df['packaging_height_cm'] = df['height__cm'] + 2  # Add some buffer for packaging
df['packaging_width_cm'] = df['width__cm'] + 2
df['packaging_depth_cm'] = df['depth__cm'] + 2

df['packaging_volume_cm3'] = df['packaging_height_cm'] * df['packaging_width_cm'] * df['packaging_depth_cm']
df['total_packaging_volume_cm3'] = df['packaging_volume_cm3'] * df['amount']

# Calculate packaging efficiency
df['packaging_efficiency_ratio'] = df['packaging_volume_cm3'] / df['item_volume_cm3']

# Calculate total volumes and efficiency
total_item_volume = df['total_item_volume_cm3'].sum()
total_packaging_volume = df['total_packaging_volume_cm3'].sum()
average_efficiency_ratio = df['packaging_efficiency_ratio'].mean()

# Group by size category
size_category_summary = df.groupby('size_category').agg(
    total_item_volume_cm3=('total_item_volume_cm3', 'sum'),
    total_packaging_volume_cm3=('total_packaging_volume_cm3', 'sum'),
    average_efficiency_ratio=('packaging_efficiency_ratio', 'mean')
)

# Print insights
print("Total item volume (cm³):", total_item_volume)
print("Total packaging volume (cm³):", total_packaging_volume)
print("Average packaging efficiency ratio:", average_efficiency_ratio)
print("\nSize category summary:")
print(size_category_summary)
```

```python
import plotly.graph_objects as go

# Load the data (assuming size_category_summary is already calculated)
df = pd.read_csv("/content/clean_data.csv")

# Function to calculate volumes and efficiency based on buffer size
def calculate_volumes_with_buffer(df, buffer):
    df['item_volume_cm3'] = df['height__cm'] * df['width__cm'] * df['depth__cm']
    df['total_item_volume_cm3'] = df['item_volume_cm3'] * df['amount']

    df['packaging_height_cm'] = df['height__cm'] + buffer
    df['packaging_width_cm'] = df['width__cm'] + buffer
    df['packaging_depth_cm'] = df['depth__cm'] + buffer

    df['packaging_volume_cm3'] = df['packaging_height_cm'] * df['packaging_width_cm'] * df['packaging_depth_cm']
    df['total_packaging_volume_cm3'] = df['packaging_volume_cm3'] * df['amount']

    df['packaging_efficiency_ratio'] = df['packaging_volume_cm3'] / df['item_volume_cm3']

    total_item_volume = df['total_item_volume_cm3'].sum()
    total_packaging_volume = df['total_packaging_volume_cm3'].sum()
    average_efficiency_ratio = df['packaging_efficiency_ratio'].mean()

    return df, total_item_volume, total_packaging_volume, average_efficiency_ratio

# Test different buffer sizes to find the optimal one
min_buffer = 0.5  # minimum buffer to avoid damage
max_buffer = 5  # maximum buffer for safety
optimal_buffer = min_buffer
optimal_total_packaging_volume = float('inf')

for buffer in range(int(min_buffer * 10), int(max_buffer * 10) + 1):
    buffer = buffer / 10.0
    df, total_item_volume, total_packaging_volume, average_efficiency_ratio = calculate_volumes_with_buffer(df, buffer)

    if total_packaging_volume < optimal_total_packaging_volume:
        optimal_total_packaging_volume = total_packaging_volume
        optimal_buffer = buffer

# Calculate final volumes with optimal buffer
df, total_item_volume, total_packaging_volume, average_efficiency_ratio = calculate_volumes_with_buffer(df, optimal_buffer)

# Group by size category with optimal buffer
size_category_summary = df.groupby('size_category').agg(
    total_item_volume_cm3=('total_item_volume_cm3', 'sum'),
    total_packaging_volume_cm3=('total_packaging_volume_cm3', 'sum'),
    average_efficiency_ratio=('packaging_efficiency_ratio', 'mean')
)

# Create the bar chart for size category summary
fig = go.Figure()

# Add total item volume bars
fig.add_trace(go.Bar(
    x=size_category_summary.index,
    y=size_category_summary['total_item_volume_cm3'],
    name='Total Item Volume',
    marker_color='rgba(50, 171, 96, 0.6)',
    text=[f'{v:,.0f} cm³' for v in size_category_summary['total_item_volume_cm3']],
    textposition='outside',
    width=0.25
))

# Add total packaging volume bars
fig.add_trace(go.Bar(
    x=size_category_summary.index,
    y=size_category_summary['total_packaging_volume_cm3'],
    name='Total Packaging Volume',
    marker_color='rgba(255, 183, 77, 0.6)',
    text=[f'{v:,.0f} cm³' for v in size_category_summary['total_packaging_volume_cm3']],
    textposition='outside',
    width=0.25
))

# Add average efficiency ratio as a line plot on secondary y-axis
fig.add_trace(go.Scatter(
    x=size_category_summary.index,
    y=size_category_summary['average_efficiency_ratio'],
    name='Average Efficiency Ratio',
    marker_color='rgba(66, 133, 244, 0.6)',
    mode='lines+markers+text',
    text=[f'{v:.2f}' for v in size_category_summary['average_efficiency_ratio']],
    textposition='top center',
    yaxis='y2'
))

# Update layout
fig.update_layout(
    title='Size Category Summary',
    xaxis_title='Size Category',
    yaxis_title='Volume (cm³)',
    yaxis2=dict(
        title='Average Efficiency Ratio',
        titlefont_size=16,
        tickfont_size=14,
        overlaying='y',
        side='right',
        range=[0, size_category_summary['average_efficiency_ratio'].max() * 1.2]
    ),
    barmode='group',
    xaxis=dict(tickfont_size=14),
    legend_title='Metrics',
    legend=dict(x=0.01, y=0.99, bgcolor='rgba(255,255,255,0.5)', bordercolor='rgba(255,255,255,0.5)'),
    template='plotly_white',
    height=600
)

# Show the figure
fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("/items 1.csv")

# Function to calculate volumes and efficiency based on buffer size
def calculate_volumes_with_buffer(df, buffer):
    df['item_volume_cm3'] = df['height__cm'] * df['width__cm'] * df['depth__cm']
    df['total_item_volume_cm3'] = df['item_volume_cm3'] * df['amount']

    df['packaging_height_cm'] = df['height__cm'] + buffer
    df['packaging_width_cm'] = df['width__cm'] + buffer
    df['packaging_depth_cm'] = df['depth__cm'] + buffer

    df['packaging_volume_cm3'] = df['packaging_height_cm'] * df['packaging_width_cm'] * df['packaging_depth_cm']
    df['total_packaging_volume_cm3'] = df['packaging_volume_cm3'] * df['amount']

    df['packaging_efficiency_ratio'] = df['packaging_volume_cm3'] / df['item_volume_cm3']

    total_item_volume = df['total_item_volume_cm3'].sum()
    total_packaging_volume = df['total_packaging_volume_cm3'].sum()
    average_efficiency_ratio = df['packaging_efficiency_ratio'].mean()

    return df, total_item_volume, total_packaging_volume, average_efficiency_ratio

# Test different buffer sizes to find the optimal one
min_buffer = 0.5  # minimum buffer to avoid damage
max_buffer = 5  # maximum buffer for safety
optimal_buffer = min_buffer
optimal_total_packaging_volume = float('inf')

for buffer in range(int(min_buffer * 10), int(max_buffer * 10) + 1):
    buffer = buffer / 10.0
    df, total_item_volume, total_packaging_volume, average_efficiency_ratio = calculate_volumes_with_buffer(df, buffer)

    if total_packaging_volume < optimal_total_packaging_volume:
        optimal_total_packaging_volume = total_packaging_volume
        optimal_buffer = buffer

# Calculate final volumes with optimal buffer
df, total_item_volume, total_packaging_volume, average_efficiency_ratio = calculate_volumes_with_buffer(df, optimal_buffer)

# Insights for the most wasteful categories
wasteful_categories = df.groupby('size_category').apply(lambda x: (x['total_packaging_volume_cm3'].sum() - x['total_item_volume_cm3'].sum()) / x['total_packaging_volume_cm3'].sum()).reset_index()
wasteful_categories.columns = ['size_category', 'waste_ratio']
wasteful_categories = wasteful_categories.sort_values(by='waste_ratio', ascending=False)

# Create a stylish and impressive bar plot
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# Create the barplot with Seaborn for better aesthetics
barplot = sns.barplot(x='waste_ratio', y='size_category', data=wasteful_categories, palette='viridis')

# Add annotations to the bars
for index, value in enumerate(wasteful_categories['waste_ratio']):
    barplot.text(value, index, f'{value:.2%}', color='black', ha="left", va="center", fontweight='bold')

# Customize the plot
plt.title('Waste Ratio by Size Category', fontsize=20, fontweight='bold', color='navy')
plt.xlabel('Waste Ratio', fontsize=16, fontweight='bold', color='darkred')
plt.ylabel('Size Category', fontsize=16, fontweight='bold', color='darkred')
plt.xticks(fontsize=12, color='darkgreen')
plt.yticks(fontsize=12, color='darkgreen')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
